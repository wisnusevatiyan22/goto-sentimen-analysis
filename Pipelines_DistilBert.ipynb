```python
from sklearn.metrics import confusion_matrix
import numpy as np
# Filter the DataFrame to keep only positive sentiment
df_positive = df[df['Sentiment_Label'] == 'positive']

# Tokenize the text
model_name = "lxyuan/distilbert-base-multilingual-cased-sentiments-student"
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertForSequenceClassification.from_pretrained(model_name)

def preprocess_text(text):
    return text.lower()

texts = [preprocess_text(text) for text in df_positive['Headline']]
labels = [2] * len(texts)  # Set labels to 2 for positive sentiment

encoding = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
input_ids = encoding['input_ids']
attention_mask = encoding['attention_mask']

# Perform inference with the model
with torch.no_grad():
    outputs = model(input_ids=input_ids, attention_mask=attention_mask)

logits = outputs.logits
predicted_labels = torch.argmax(logits, dim=1)

# Convert predicted_labels to NumPy array
predicted_labels = predicted_labels.cpu().numpy()

# Create a confusion matrix
confusion = confusion_matrix(labels, predicted_labels)

# Define class labels for the confusion matrix
class_labels = ['negative', 'neutral', 'positive']

# Print the confusion matrix
print("Confusion Matrix:")
print(confusion)

# Calculate accuracy, precision, recall, and F1-score
accuracy = accuracy_score(labels, predicted_labels)
precision = np.diag(confusion) / np.sum(confusion, axis=0)
recall = np.diag(confusion) / np.sum(confusion, axis=1)
f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python

# Mengubah kolom 'Date' menjadi tipe datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d %b %Y', errors='coerce')

# Membuat DataFrame dengan tanggal yang belum tersedia
date_range = pd.date_range(start=df['Date'].min(), end=df['Date'].max())
missing_dates = set(date_range) - set(df['Date'])

missing_data = {
    'Date': list(missing_dates),
    'Headline': ['tidak terdapat berita'] * len(missing_dates),
    'Sentiment_Label': ['neutral'] * len(missing_dates),
    'Sentiment_Score': [0.0] * len(missing_dates)
}

missing_df = pd.DataFrame(missing_data)

# Menggabungkan DataFrame utama dengan DataFrame yang berisi tanggal yang belum tersedia
final_df = pd.concat([df, missing_df])

# Mengurutkan DataFrame berdasarkan kolom 'Date'
final_df = final_df.sort_values(by='Date').reset_index(drop=True)
```

```python
final_df.tail()
```

```python
final_df = final_df.dropna(subset=['Date'])
```

```python
final_df.tail()
```

```python
import numpy as np

# Buat array berisi nama-nama hari dalam seminggu
days_of_week = np.array(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])

# Tambahkan kolom 'Day_of_Week' ke DataFrame untuk menentukan hari dalam seminggu
final_df['Day_of_Week'] = final_df['Date'].dt.day_name()

# Saring DataFrame untuk hanya menyertakan data pada hari kerja
final_df = final_df[final_df['Day_of_Week'].isin(days_of_week)]

# Hapus kolom 'Day_of_Week' jika tidak lagi diperlukan
final_df = final_df.drop(columns=['Day_of_Week'])

# Reset indeks DataFrame
final_df = final_df.reset_index(drop=True)
```
